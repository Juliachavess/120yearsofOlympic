# **Dados faltantes ou nulos**


```python
# dropna apaga todas as linhas que tenham dados faltantes
dados2 = dados.dropna()
```

```python
dados2.head()
```

```python
#perdeu muita informação 
dados2.shape
```

```python
# se for nulo vai estar True
nulos = dados.isnull()
nulos
```

```python
# somando todos os valores nulos
faltantes = dados.isnull().sum()
faltantes
```

```python
# calculando a porcntagem de dados faltantes de cada coluna
faltantes_porcent = (faltantes / len(dados['ID']))*100
faltantes_porcent
```

```python
# substituir os valores nulos
dados['Medal'].fillna('Nenhuma', inplace = True)
dados['Age'].fillna(dados['Age'].mean(), inplace = True)
dados['Altura'].fillna(dados['Altura'].mean(), inplace = True)
dados['Peso'].fillna(dados['Peso'].mean(), inplace = True)
dados
```
